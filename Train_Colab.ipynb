## 3. Training
We train for 50 epochs. The model will learn to detect:
- 0: Fire
- 1: Smoke

```python
from ultralytics import YOLO

# Load pre-trained model
model = YOLO("yolov8n.pt")

# Train
results = model.train(
    data="./datasets/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    name="fire_smoke_model"
)
```
